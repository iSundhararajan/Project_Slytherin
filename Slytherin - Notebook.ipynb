```python
import random
import string
def generate_unique_id(existing_ids, length=5):
    while True:
        random_id = ''.join(random.choices(string.ascii_letters + string.digits, k=length))
        if random_id not in existing_ids:
            return random_id
existing_ids = []
unique_ids = []
for _ in range(len(merged_df_2)):
    new_id = generate_unique_id(existing_ids)
    unique_ids.append(new_id)
merged_df_2['person_id'] = unique_ids
```

```python
unique_ids = []
for _ in range(len(merged_df_2)):
    new_id = generate_unique_id(existing_ids,3)
    unique_ids.append(new_id)
    existing_ids.append(new_id)
```

```python
merged_df_2['HP_id'] = unique_ids
```
